```python
from google.colab import drive
import pandas as pd
import sqlite3
import os
from tabulate import tabulate

drive.mount('/content/gdrive')
caminho_banco = '/content/gdrive/MyDrive/bancoSQL/'
nomedodataset = 'dataset_bancomundial_paises (1).csv'

if not os.path.exists(caminho_banco):
    os.makedirs(caminho_banco)
    print("Pasta 'bancoSQL' criada com sucesso!")

caminho_arquivo_csv = '/content/gdrive/MyDrive/' + nomedodataset
dados = pd.read_csv(caminho_arquivo_csv, sep=';', decimal=',')
caminho_banco_dados = caminho_banco + 'dataset_bancomundial.db'

def salvar_dados(dados, nome_tabela, caminho_banco_dados):
    conn = sqlite3.connect(caminho_banco_dados)
    if conn.execute(f"SELECT name FROM sqlite_master WHERE type='table' AND name='{nome_tabela}'").fetchone():
        conn.execute(f"DROP TABLE {nome_tabela}")
    dados.to_sql(nome_tabela, conn, index=False)
    conn.close()
    print(f"Tabela '{nome_tabela}' foi criada e os dados foram salvos com sucesso no banco de dados.")

nome_tabela = 'dados_bancomundial'
salvar_dados(dados, nome_tabela, caminho_banco_dados)
conn = sqlite3.connect(caminho_banco_dados)
consulta_sql = f"SELECT * FROM {nome_tabela} LIMIT 10;"
resultado = pd.read_sql_query(consulta_sql, conn)
print(tabulate(resultado, headers='keys', tablefmt='pretty'))

```

```python
from google.colab import drive
import pandas as pd
import sqlite3

# Montar o Google Drive para acessar os arquivos
drive.mount('/content/gdrive')

# Caminho do banco de dados SQLite no Google Drive
caminho_banco_dados = '/content/gdrive/MyDrive/bancoSQL/dataset_bancomundial.db'

# Conectar ao banco de dados SQLite
conn = sqlite3.connect(caminho_banco_dados)

# Comandos SQL para criar tabelas 'Country', 'Indicator' e 'Data', caso não existam
comando_cria_tabela_country = "\
    CREATE TABLE IF NOT EXISTS Country (\
        CountryCode VARCHAR PRIMARY KEY NOT NULL,\
        ShortName VARCHAR NOT NULL,\
        LongName VARCHAR NOT NULL,\
        Region VARCHAR NOT NULL\
    );"

comando_cria_tabela_indicator = "\
    CREATE TABLE IF NOT EXISTS Indicator (\
        IndicatorCode VARCHAR PRIMARY KEY NOT NULL,\
        IndicatorName VARCHAR NOT NULL\
    );"

comando_cria_tabela_data = "\
    CREATE TABLE IF NOT EXISTS Data (\
        CountryCode VARCHAR,\
        IndicatorCode VARCHAR,\
        Year INTEGER,\
        Value INTEGER,\
        FOREIGN KEY(CountryCode) REFERENCES Country(CountryCode),\
        FOREIGN KEY(IndicatorCode) REFERENCES Indicator(IndicatorCode)\
    );"

# Executar os comandos para criar as tabelas
conn.executescript(comando_cria_tabela_country)
conn.executescript(comando_cria_tabela_indicator)
conn.executescript(comando_cria_tabela_data)

# Consulta para listar todas as tabelas presentes no banco de dados
tabelas = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", conn)

# Exibir as tabelas
display(tabelas)

# Fechar a conexão com o banco de dados
conn.close()
```
