```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt

drive.mount('/content/gdrive')
caminho='/content/gdrive/MyDrive/'
conjunto_de_dados='ds_chuva_sm_2011_2020_10.csv'
caminho_conjunto_de_dados=caminho+conjunto_de_dados

# Carregar os dados
df = pd.read_csv(caminho_conjunto_de_dados, sep=';', decimal=',')
df['DT_MEDICAO'] = pd.to_datetime(df['DT_MEDICAO'], format='%Y-%m-%d')

# Filtrar dados para o ano de 2015
df_2015 = df[df['DT_MEDICAO'].dt.year == 2015]

# Calcular a precipitação acumulada para cada mês
chuva_acumulada_2015 = df_2015.groupby(df_2015['DT_MEDICAO'].dt.month)['CHUVA'].sum()

# Plotar histograma
plt.figure(figsize=(10,6))
plt.bar(chuva_acumulada_2015.index, chuva_acumulada_2015.values)
plt.xlabel('Mês')
plt.ylabel('Precipitação Acumulada')
plt.title('Precipitação Acumulada para Cada Mês em 2015')
plt.grid(True)
plt.show()
```
