```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt

def carregar_dados(caminho_conjunto_de_dados):
    df = pd.read_csv(caminho_conjunto_de_dados, sep=';', decimal=',', parse_dates=['DT_MEDICAO'])
    return df

def analisar_e_plotar_precipitacao_ano(df, ano):
    df_ano = df[df['DT_MEDICAO'].dt.year == ano]
    chuva_acumulada = df_ano.groupby(df_ano['DT_MEDICAO'].dt.month)['CHUVA'].sum()

    plt.figure(figsize=(10, 6))
    plt.bar(chuva_acumulada.index, chuva_acumulada.values)
    plt.xlabel('Mês')
    plt.ylabel('Precipitação Acumulada')
    plt.title(f'Precipitação Acumulada para Cada Mês em {ano}')
    plt.grid(True)
    plt.xticks(range(1, 13), ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'])
    plt.show()

def main():
    caminho = '/content/gdrive/MyDrive/'
    conjunto_de_dados = 'ds_chuva_sm_2011_2020_10.csv'
    caminho_conjunto_de_dados = f'{caminho}{conjunto_de_dados}'

    df = carregar_dados(caminho_conjunto_de_dados)
    ano = 2018
    analisar_e_plotar_precipitacao_ano(df, ano)

if __name__ == "__main__":
    main()
```
